```python
import skimage.io
from skimage.filters import unsharp_mask

image = skimage.io.imread('D:\\MULANGA\\myproject\\images2\\phot5.jpeg')

sharpened_image_1 = unsharp_mask(image, radius=1, amount=1)
sharpened_image_2 = unsharp_mask(image, radius=5, amount=2)
sharpened_image_3 = unsharp_mask(image, radius=10, amount=3)

skimage.io.imshow_collection([image, sharpened_image_1, sharpened_image_2, sharpened_image_3])
skimage.io.show()
```

```python
from PIL import Image, ImageFilter

# Open the image
image = Image.open('D:\\MULANGA\\myproject\\photos\\gate.jpeg')

# Apply UnsharpMask filter with different parameters
sharpened_image_1 = image.filter(ImageFilter.UnsharpMask(radius=2, percent=150))
sharpened_image_2 = image.filter(ImageFilter.UnsharpMask(radius=5, percent=200))
sharpened_image_3 = image.filter(ImageFilter.UnsharpMask(radius=10, percent=250))

# Display the original and sharpened images
image.show()
sharpened_image_1.show()
sharpened_image_2.show()
sharpened_image_3.show()
```

```python
import skimage.io
import skimage.transform
import matplotlib.pyplot as plt

# Load the Lena grayscale image
image = skimage.io.imread('D:\\MULANGA\\myproject\\photos\\gate.jpeg', as_gray=True)

# Generate the Gaussian pyramid
pyramid = skimage.transform.pyramid_laplacian(image, downscale=2, max_layer=5)

# Display the pyramid
fig, axes = plt.subplots(nrows=1, ncols=8, figsize=(12, 3))

# Display the original image
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Level 0')

# Display the pyramid levels
for i, layer in enumerate(pyramid, start=1):
    axes[i].imshow(layer, cmap='gray')
    axes[i].set_title('Level {}'.format(i))

# Hide the empty subplots
for j in range(5, 0, -1):
    axes[-j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import skimage.io
import skimage.transform
import matplotlib.pyplot as plt

# Load the Lena grayscale image
image = skimage.io.imread('D:\\MULANGA\\myproject\\photos\\matching.jpg', as_gray=True)

# Generate the Gaussian pyramid
gaussian_pyramid = skimage.transform.pyramid_gaussian(image, downscale=2)

# Display the Gaussian pyramid
fig, axes = plt.subplots(figsize=(8, 6))

# Display the original image
axes.imshow(image, cmap='gray')
axes.set_title('Level 0')
axes.axis('off')

# Display the pyramid levels
level = 1
for layer in gaussian_pyramid:
    axes = fig.add_subplot(8, 6, level)
    axes.imshow(layer, cmap='gray')
    axes.set_title('Level {}'.format(level))
    axes.axis('off')
    level += 1

plt.tight_layout()
plt.show()
```
